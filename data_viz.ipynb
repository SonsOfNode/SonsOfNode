## Exemplo 1: Gráfico de Dispersão Interativo

```python
import pandas as pd

# Crie um dicionário de mapeamento com os critérios de idade
idade_mapping = {
    1: 'Menor de 17 anos',
    2: '17 anos',
    3: '18 anos',
    4: '19 anos',
    5: '20 anos',
    6: '21 anos',
    7: '22 anos',
    8: '23 anos',
    9: '24 anos',
    10: '25 anos',
    11: 'Entre 26 e 30 anos',
    12: 'Entre 31 e 35 anos',
    13: 'Entre 36 e 40 anos',
    14: 'Entre 41 e 45 anos',
    15: 'Entre 46 e 50 anos',
    16: 'Entre 51 e 55 anos',
    17: 'Entre 56 e 60 anos',
    18: 'Entre 61 e 65 anos',
    19: 'Entre 66 e 70 anos',
    20: 'Maior de 70 anos'
}

# Substitua os valores na coluna 'gp_idade' usando o dicionário de mapeamento
df['gp_idade'] = df['gp_idade'].replace(idade_mapping)
```

```python
ordem_idade = [
    'Menor de 17 anos',
    '17 anos',
    '18 anos',
    '19 anos',
    '20 anos',
    '21 anos',
    '22 anos',
    '23 anos',
    '24 anos',
    '25 anos',
    'Entre 26 e 30 anos',
    'Entre 31 e 35 anos',
    'Entre 36 e 40 anos',
    'Entre 41 e 45 anos',
    'Entre 46 e 50 anos',
    'Entre 51 e 55 anos',
    'Entre 56 e 60 anos',
    'Entre 61 e 65 anos',
    'Entre 66 e 70 anos',
    'Maior de 70 anos'
]

# Aplica a ordem à coluna gp_idade
df['gp_idade'] = pd.Categorical(df['gp_idade'], categories=ordem_idade, ordered=True)

# Ordena o DataFrame pela coluna gp_idade
df = df.sort_values('gp_idade')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['gp_idade'], df['nota_mat'], alpha=0.5)
plt.title('Idade vs. Nota de Matemática')
plt.xlabel('Idade')
plt.ylabel('Nota de Matemática')
plt.xticks(rotation=90)
plt.show()
```

```python
# Função para mapear faixas etárias para valores numéricos
def mapear_idade(idade_str):
    if idade_str == 'Menor de 17 anos':
        return 1
    elif idade_str == '17 anos':
        return 2
    elif idade_str == '18 anos':
        return 3
    elif idade_str == '19 anos':
        return 4
    elif idade_str == '20 anos':
        return 5
    elif idade_str == '21 anos':
        return 6
    elif idade_str == '22 anos':
        return 7
    elif idade_str == '23 anos':
        return 8
    elif idade_str == '24 anos':
        return 9
    elif idade_str == '25 anos':
        return 10
    elif idade_str == 'Entre 26 e 30 anos':
        return 11
    elif idade_str == 'Entre 31 e 35 anos':
        return 12
    elif idade_str == 'Entre 36 e 40 anos':
        return 13
    elif idade_str == 'Entre 41 e 45 anos':
        return 14
    elif idade_str == 'Entre 46 e 50 anos':
        return 15
    elif idade_str == 'Entre 51 e 55 anos':
        return 16
    elif idade_str == 'Entre 56 e 60 anos':
        return 17
    elif idade_str == 'Entre 61 e 65 anos':
        return 18
    elif idade_str == 'Entre 66 e 70 anos':
        return 19
    elif idade_str == 'Maior de 70 anos':
        return 20

# Aplica a função para criar uma nova coluna 'faixa_etaria'
df['faixa_etaria'] = df['gp_idade'].apply(mapear_idade)

# Agrupa por faixa etária e calcula estatísticas resumidas
faixa_etaria = df.groupby('faixa_etaria').agg({
    'nota_cien': 'mean',
    'nota_hum': 'mean',
    'nota_port': 'mean',
    'nota_mat': 'mean'
}).reset_index()

# Exibe o resultado
print(faixa_etaria)
```

```python
import numpy as np

# Lista de estados civis
estados_civis = df['est_civil'].unique()

# Definindo as cores para cada estado civil
cores = ['blue', 'red', 'green', 'purple', 'orange']

# Criando uma lista para as idades
idades = ['Menor de 17 anos', '17 anos', '18 anos', '19 anos', '20 anos', '21 anos', '22 anos', '23 anos', 
          '24 anos', '25 anos', 'Entre 26 e 30 anos', 'Entre 31 e 35 anos', 'Entre 36 e 40 anos', 
          'Entre 41 e 45 anos', 'Entre 46 e 50 anos', 'Entre 51 e 55 anos', 'Entre 56 e 60 anos', 
          'Entre 61 e 65 anos', 'Entre 66 e 70 anos', 'Maior de 70 anos']

# Criando um índice para as idades
x = np.arange(len(idades))

# Largura das barras
largura = 0.15

# Criando o gráfico
plt.figure(figsize=(12, 8))

for i, est_civil in enumerate(estados_civis):
    # Filtrando o DataFrame para cada estado civil
    df_est_civil = df[df['est_civil'] == est_civil]
    
    # Calculando a média da nota de Matemática para cada faixa etária
    notas_medias = df_est_civil.groupby('faixa_etaria')['nota_mat'].mean()
    
    # Plotando as barras
    plt.bar(x + (largura * i), notas_medias, largura, label=f'Estado Civil {est_civil}', color=cores[i])

# Adicionando rótulos
plt.xlabel('Faixa Etária')
plt.ylabel('Nota de Matemática')
plt.title('Média da Nota de Matemática por Faixa Etária e Estado Civil')
plt.xticks(x + (largura * 2), idades, rotation=90)
plt.legend(title='Estado Civil')

plt.tight_layout()
plt.show()
```
