## Valores Ausentes

Vamos conferir agora o número total de linhas de dados que temos em nossa Dataframe antes de que precisemos pensar em fazer qualquer tipo de remoção.

```python
# Conferindo as dimensões da tabela
df.shape
```

Podemos observar que estamos lidando com um conjunto considerável de dados. Foram mais de 3 milhões de estudantes cadastrados para a realização da prova do ENEM no ano de 2022.

Vamos agora criar uma pequena visualização de quantos valores ausentes temos por coluna e qual porcentagem eles representam em relação a todos os dados desta mesma coluna:

```python
# Criando uma variável que armazena a quantidade de valores ausentes por coluna
num_valores_ausentes = df.isna().sum()

# Criando uma variável que armazena a porcentagem de valores ausentes para cada coluna
porcentagem_ausente = (df.isna().mean() * 100).round(2)

# Criando uma tabela que evidencie a soma e a porcentagem de NaN por coluna
df_ausentes = pd.DataFrame({
    'Num. de Valores Ausentes': num_valores_ausentes,
    'Porcentagem Ausente': porcentagem_ausente
})

print(df_ausentes)
```

```python
# Preenchendo os valores ausentes na coluna 'tp_ensino' com 0
# indicando os caso em que as necessidades especiais não foram informadas
df['tp_ensino'] = df['tp_ensino'].fillna(0)
```

Como podemos observar em nosso dicionário acima, existeram alunos que faltaram em determinadas provas ou foram eliminados delas. Provavelmente estes podem ser responsáveis pela alta presença (entre 28 e 32%) de valores NaN nas notas das provas.

Vamos substituir estes valores por zeros, já que numericamente eles não terão mais do que esse valor em pontuação.

```python
# Para as linhas onde o participante faltou (0) ou foi eliminado (2) na prova de Ciências da Natureza,
# a nota correspondente é definida como 0 em 'nota_cien'
df.loc[(df['present_cien'] == 0) | (df['present_cien'] == 2), 'nota_cien'] = 0
```

```python
# Para as linhas onde o participante faltou (0) ou foi eliminado (2) na prova de Ciências Humanas,
# a nota correspondente é definida como 0 em 'nota_hum'
df.loc[(df['present_hum'] == 0) | (df['present_hum'] == 2), 'nota_hum'] = 0
```

```python
# Para as linhas onde o participante faltou (0) ou foi eliminado (2) na prova de Linguagens e Códigos,
# a nota correspondente é definida como 0 em 'nota_port'
df.loc[(df['present_port'] == 0) | (df['present_port'] == 2), 'nota_port'] = 0
```

```python
# Para as linhas onde o participante faltou (0) ou foi eliminado (2) na prova de Matemática,
# a nota correspondente é definida como 0 em 'nota_mat'
df.loc[(df['present_mat'] == 0) | (df['present_mat'] == 2), 'nota_mat'] = 0
```

Vamos conferir mais uma vez a presença de valores ausentes em nossos dados:

```python
num_valores_ausentes = df.isna().sum()

porcentagem_ausente = (df.isna().mean() * 100).round(2)

df_ausentes = pd.DataFrame({
    'Num. de Valores Ausentes': num_valores_ausentes,
    'Porcentagem Ausente': porcentagem_ausente
})

print(df_ausentes)
```
